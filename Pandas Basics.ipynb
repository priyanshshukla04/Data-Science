# Merging DataFrames in pandas

```python
df2 = pd.DataFrame([[1,3,45,2],[112,3,55,6],[12,2,3,4],[23,55,77,88],[12,456,7,8]],columns = ['A','B','C','D'])
```

```python
df2
```

```python
df3 = pd.DataFrame([[11,3,452,21],[12,3,55,60],[2,2,23,14],[223,55,7,88],[12,456,7,80]],columns = ['X','Y','Z','W'])
```

```python
df3
```

```python
df4 = pd.merge(df2, df3, left_on = 'B', right_on = 'Y')
```
